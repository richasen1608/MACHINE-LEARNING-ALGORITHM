```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df=pd.read_csv("housepricing.csv")
df
```

```python
import math
median_room=math.floor(df.bedroom.median())
median_room
```

```python
df.bedroom=df.bedroom.fillna(median_room)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area','bedroom','age']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```
